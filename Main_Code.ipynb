````python
EmploymentDatadf['observation_date'] = pd.to_datetime(EmploymentDatadf['observation_date'])
PatentDatadf['pub_dt'] = pd.to_datetime(PatentDatadf['pub_dt'])
ManagementEmploymentDatadf['observation_date'] = pd.to_datetime(ManagementEmploymentDatadf['observation_date'])
ServicesEmploymentDatadf['observation_date'] = pd.to_datetime(ServicesEmploymentDatadf['observation_date'])
EngineeringEmploymentDatadf['observation_date'] = pd.to_datetime(EngineeringEmploymentDatadf['observation_date'])
GDPGrowthDatadf['observation_date'] = pd.to_datetime(GDPGrowthDatadf['observation_date'])

# Create a new 'month' column in each dataset to store the month
EmploymentDatadf['month'] = EmploymentDatadf['observation_date'].dt.to_period('M')
PatentDatadf['month'] = PatentDatadf['pub_dt'].dt.to_period('M')
ManagementEmploymentDatadf['month'] = ManagementEmploymentDatadf['observation_date'].dt.to_period('M')
ServicesEmploymentDatadf['month'] = ServicesEmploymentDatadf['observation_date'].dt.to_period('M')
EngineeringEmploymentDatadf['month'] = EngineeringEmploymentDatadf['observation_date'].dt.to_period('M')
GDPGrowthDatadf['month'] = GDPGrowthDatadf['observation_date'].dt.to_period('M')

# Merge the datasets based on the 'month' column
merged_df = EmploymentDatadf.merge(PatentDatadf, on='month', how='inner')
merged_df = merged_df.merge(ManagementEmploymentDatadf, on='month', how='inner')
merged_df = merged_df.merge(ServicesEmploymentDatadf, on='month', how='inner', suffixes=('_left', '_right'))
merged_df = merged_df.merge(EngineeringEmploymentDatadf, on='month', how='inner')
merged_df = merged_df.merge(GDPGrowthDatadf, on='month', how='inner')

# Save the merged dataset to a new CSV file
merged_df.to_csv(Patentspath + '/'+'merged_dataset_by_month.xlsx', index=False)
````

````python
merged_df.head()
````

````python
count_patents = merged_df.groupby(by="month", dropna=False).sum()
count_patents.reset_index(inplace=True)
count_patents.rename(columns={"year": "count_patents"}, inplace=True)
count_patents = count_patents[["month", "count_patents"]]
count_patents.tail(50)
count_patents.head()



dfCount = merged_df.merge(count_patents, on='month', how='inner')
dfCount.tail()

#count by type of patent, use predict50 columns
#merge different counts of patents with dfCount
#try it with the different ys
````

# Models

````python
count_mlpatents = merged_df.groupby(by="month", dropna=False).sum()
count_mlpatents.reset_index(inplace=True)
count_mlpatents = count_mlpatents[["month", "predict50_ml"]]
count_mlpatents.tail(50)
count_mlpatents.head()

dfCount = dfCount.merge(count_mlpatents, on='month', how='inner')

dfCount.rename(columns={"predict50_ml_y": "count_patents_ml"}, inplace=True)

dfCount.tail()
````

````python

count_nlppatents = merged_df.groupby(by="month", dropna=False).sum()
count_nlppatents.reset_index(inplace=True)
count_nlppatents = count_nlppatents[["month", "predict50_nlp"]]
count_nlppatents.tail(50)
count_nlppatents.head()


dfCount = dfCount.merge(count_nlppatents, on='month', how='inner')

dfCount.rename(columns={"predict50_nlp_y": "count_patents_nlp"}, inplace=True)

dfCount.tail()
````

````python
count_evopatents = merged_df.groupby(by="month", dropna=False).sum()
count_evopatents.reset_index(inplace=True)
count_evopatents = count_evopatents[["month", "predict50_evo"]]
count_evopatents.tail(50)
count_evopatents.head()
dfCount = dfCount.merge(count_evopatents, on='month', how='inner')

dfCount.rename(columns={"predict50_evo_y": "count_patents_evo"}, inplace=True)

dfCount.tail()
````

````python
count_hardwarepatents = merged_df.groupby(by="month", dropna=False).sum()
count_hardwarepatents.reset_index(inplace=True)
count_hardwarepatents = count_hardwarepatents[["month", "predict50_hardware"]]
count_hardwarepatents.tail(50)
count_hardwarepatents.head()
dfCount = dfCount.merge(count_hardwarepatents, on='month', how='inner')

dfCount.rename(columns={"predict50_hardware_y": "count_patents_hardware"}, inplace=True)

dfCount.tail()
````

````python
X = dfCount[['count_patents_ml', 'Monthly Real GDP Index']].values
y1 = dfCount['CES6056140001']
y2 = dfCount['CES6055000001']
y3 = dfCount['CES6054130001']
y4 = dfCount['CES6054150001']

from sklearn import datasets, linear_model, metrics

#y1 model
reg = linear_model.LinearRegression()
reg.fit(X, y1)
print('Coefficients services employment: ', reg.coef_)
print('Variance score services employment: {}'.format(reg.score(X, y1)))
print('Intercept services employment: ', reg.intercept_)


#y2 model
reg = linear_model.LinearRegression()
reg.fit(X, y2)
print("")
print('Coefficients Management employment: ', reg.coef_)
print('Variance score management employment: {}'.format(reg.score(X, y2)))
print('Intercept management employment: ', reg.intercept_)

#y3 model
reg = linear_model.LinearRegression()
reg.fit(X, y3)
print("")
print('Coefficients engineering employment: ', reg.coef_)
print('Variance score engineering employment: {}'.format(reg.score(X, y3)))
print('Intercept engineering employment: ', reg.intercept_)

#y4 model
reg = linear_model.LinearRegression()
reg.fit(X, y4)
print("")
print('Coefficients employment data: ', reg.coef_)
print('Variance score employment data: {}'.format(reg.score(X, y4)))
print('Intercept employment data: ', reg.intercept_)
````

````python
from sklearn.neural_network import MLPRegressor

regr = MLPRegressor(random_state=1, max_iter=500, hidden_layer_sizes=(10,10,11)).fit(X, y1)


print('Variance score services employment: {}'.format(regr.score(X, y1)))
````

````python
regr = MLPRegressor(random_state=1, max_iter=500, hidden_layer_sizes=(9,10,12)).fit(X, y2)


print('Variance score management employment: {}'.format(regr.score(X, y2)))
````

````python
regr = MLPRegressor(random_state=1, max_iter=500, hidden_layer_sizes=(11,12,10)).fit(X, y3)


print('Variance score engineering employment: {}'.format(regr.score(X, y3)))
````

````python
regr = MLPRegressor(random_state=1, max_iter=500, hidden_layer_sizes=(8,10,11)).fit(X, y4)


print('Variance score overall employment: {}'.format(regr.score(X, y4)))
````
